```python
import requests
from bs4 import BeautifulSoup
url = 'https://www.104.com.tw/jobs/search/?'
url = 'https://www.104.com.tw/jobs/search/'
para = {
    'keyword': '爬蟲',
    'order': '1',
    'jobsource' : '2018indexpoc',
    'ro' : '1'
    
}
request_url = "https://www.104.com.tw/jobs/search/"
response = requests.get(url, params=para)
soup = BeautifulSoup(response.text,"lxml")

number = 1
for i in soup.find_all('article' , class_ = 'b-block--top-bord job-list-item b-clearfix js-job-item'):
    print('項目 : '  , number)
    print('刊登日期 : ' , i.select_one('span' , class_ = 'b-tit__date').text)
    print('職稱 : ' , i.get('data-job-name'))
    print('待遇 : ' , i.find('span' , class_ = 'b-tag--default').text)
    print('網址 : ', 'http://' + i.select('a')[0]['href']) 
    print('公司名稱 : ' , i.get('data-cust-name'))
    print('應徵人數 :' , i.find('a' , class_ = 'b-link--gray').text.rstrip('應徵'))
    print('-------------------------------')
    number += 1
```
