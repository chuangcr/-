```python
import requests
from bs4 import BeautifulSoup
import json
website = 'https://www.104.com.tw/company/ajax/joblist/1a2x6bi34e?roleJobCat=0_0&area=0&page=1&pageSize=20&order=8&asc=0'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.54 Safari/537.36'
          ,'Referer': 'https://www.104.com.tw/company/1a2x6bi34e?roleJobCat=0_0&area=0&page=1&pageSize=20&order=8&asc=0&tab=job',}
try: 
    content  = requests.get(website,headers=headers)
except:
    print('網址錯誤')
    
data = BeautifulSoup(content.text)
data = json.loads(data.text)
num = 1
for i in data['data']['list']['normalJobs']:
    print('項目 : ' , num)
    print('刊登日期 :' , i['appearDate'])
    print('職稱 : ' , i['jobName'])
    print('工作地區 : ' , i['jobAddrNoDesc'])
    print('待遇 : ' , i['jobTags'])
    print('--------------------------')
    num += 1
```
